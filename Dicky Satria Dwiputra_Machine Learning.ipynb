## Customer Segmentation

In this process, we will try to segment or group the customers based on 3 characteristics. The characteristics that will be considered are:
1. Recency : Recency of customer orders
2. Frequency: The amount of times the customer has ordered
3. Monetary: The amount of money the customer has spent

```python
df2.head()
```

###### Recency

```python
df2['Tanggal_Transaksi'] = pd.to_datetime(df2['Tanggal_Transaksi'])
```

Now we will define the recency calulation formula that is the difference in days between the overall latest transaction time and the latest transaction time for each customer (country).

```python
latest = max(df2.Tanggal_Transaksi)
df2['time_range'] = latest - df2['Tanggal_Transaksi']
recent = df2.groupby('Negara').agg({'time_range': 'min'}).astype('timedelta64[D]')
recent.head()
```

###### Frequency

For Frequency, we will look at the number of times a customer has ordered.

```python
frequency = df2.groupby('Negara').agg({'Barang':'count'})
frequency.sort_values('Barang',ascending=False)
```

###### Monetary

Monetary considers the total amount of money a customer has spent on all items.

```python
monetary = df2.groupby('Negara').agg({'Harga_Satuan': 'sum'}).astype('int64')
monetary.head()
```

After defining the 3 variables used we will now combine all 3 into one dataframe and look at its descriptive statistics

```python
rfm = pd.concat([recent, frequency, monetary], axis=1)
rfm.columns = ['recency', 'frequency', 'monetary']
```

We will now give an RFM scoring to each data. All variables are given a scoring system of 1 to 5 with 5â€“5â€“5 being the best customer and 1â€“1â€“1 being the least valuable.

```python
label1 = [5, 4, 3, 2, 1]
label2 = [1, 2, 3, 4, 5]
rfm['R'] = pd.qcut(rfm['recency'],
                    q=[0, 0.2, 0.4, 0.6, 0.8, 1],
                    labels=label1)
rfm['F'] = pd.cut(rfm['frequency'], bins=[0, 1000, 5000, 10000, 100000, 500000], labels=label2)
rfm['M'] = pd.qcut(rfm['monetary'],
                    q=[0, 0.2, 0.4, 0.6, 0.8, 1],
                    labels=label2)

rfm
```

Next we will use the Isolation Forest method which classifies whether a data point is considered as an outlier or not an outlier

```python
clf = IsolationForest(random_state=50)
clf.fit(rfm)
y_pred_train = clf.predict(rfm)
y_pred_train
```

```python
rfm['Outlier']=y_pred_train
rfm['Outlier'].value_counts()
```

```python
sns.countplot(y_pred_train)
plt.title('Number of Outliers and Nonoutliers by Isolation Forest') 
```

Based on the graph, we see that there is 4 data which is considererd an outlier.

We will keep the outliers since it probably can provide us insightful information

```python
rfm2=rfm.iloc[:,3:6]
rfm2
```

We will now choose the appropriate k used for the clustering algorithm. We will use the elbow method for choosing the appropriate k

```python
distorsions, k1, kN = [], 2, 10 
for k in range(k1, kN):
    kmeans = cluster.KMeans(n_clusters=k,init='k-means++').fit(rfm2)
    distorsions.append(kmeans.inertia_)
plt.plot(range(k1, kN), distorsions); plt.grid(True)
plt.title('Elbow curve')
plt.xlabel('Value of k') 
```

As seen from the above, we choose k based on where the decrement in inertia is insignificant, so we choose k = 4

```python
k=4
kmPP = cluster.KMeans(n_clusters=k, init='k-means++', n_jobs=-1, random_state = 99)
kmPP.fit(rfm2)
C_kmpp = kmPP.predict(rfm2)
sns.countplot(C_kmpp)
C_kmpp[:10]
plt.title('Number of Data in each clusters by k-means++') 
```

```python
kmPP.cluster_centers_
```

```python
cols = ['recency','frequency','monetary']
dfC = pd.DataFrame(kmPP.cluster_centers_, columns=cols)
dfC['cluster'] = dfC.index
```

```python
freq=rfm
freq=freq.reset_index()
a=pd.DataFrame(data=C_kmpp,columns=['cluster'])
cluster=pd.merge(freq, a, left_index=True, right_index=True)
cluster
```

```python
cluster.groupby('cluster').agg({
    'recency':['mean','min','max'],
    'frequency':['mean','min','max'],
    'monetary':['mean','min','max'],
    'Negara':'count'
})
```

```python
pd.plotting.parallel_coordinates(dfC, 'cluster', color=('r', 'g', 'b','y'))
plt.grid(b=None, which='major', axis='both',color='gray')
plt.show()
```

As we can see from the graph above, the customers can be separated into 4 cluster that is:

1. Cluster 0 which has a high recency but low monetary.
2. Cluster 1 which has a high recency and high monetary. 
3. Cluster 2 which has low recency and low monetary.
4. Cluster 3 which has low recency but high monetary.
